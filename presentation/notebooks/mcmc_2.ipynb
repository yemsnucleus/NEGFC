## Loading Results

```python
import pickle
with open('./presentation/results/mcmc/etatelb/H_2/MCMC_results', 'rb') as fileSave:
    output = pickle.load(fileSave)
```

```python
chain_radius = output['chain'][..., 0]
chain_theta = output['chain'][..., 1]
chain_flux = output['chain'][..., 2]

chain_radius.shape
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8,5), sharex=True, gridspec_kw={'hspace': 0.1, 'wspace': 0.5})
for nwalk in range(6):
    axes[0].plot(chain_radius[nwalk], marker='', linestyle='-', linewidth=1, color='purple')
    axes[1].plot(chain_theta[nwalk], marker='', linestyle='-', linewidth=1, color='purple')
    axes[2].plot(chain_flux[nwalk], marker='', linestyle='-', linewidth=1, color='purple')
```
